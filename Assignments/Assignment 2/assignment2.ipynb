### 1.Email Validator

```python
email = input("Enter email: ")

if "@" in email and (email.endswith(".com") or email.endswith(".org")):
    print("Valid Email")
else:
    print("Your email is not valid!")
```

### 2.Count Word Frequency in Customer Reviews

```python
text = input("Enter Text: ").strip('"').lower().split()

unique_word = []

for i in text:
    if i not in unique_word:
        unique_word.append(i)

print("{",end="")
for i in unique_word:
    if i == unique_word[-1]:
        print(f"'{i}': {text.count(i)}", end="")
    else:
        print(f"'{i}': {text.count(i)}", end=", ")
print("}")

```

### 3. Password Strength Checker

```python
password = input("Enter password: ").strip('"')
uppercase = lowercase = digit = special = False

special_characters = r"!@#$%^&*()-_=+[]{};:'\",.<>/?\\|`~"

for i in password:
    if uppercase == False:
        uppercase = i.isupper()
    if lowercase == False:
        lowercase = i.islower()
    if digit == False:
        digit = i.isdigit()
    if special == False:
        special = i in special_characters

if uppercase and lowercase and digit and special:
    print("Strong Password")
else:
    print("Weak Password")
# "Hello@123"
print(password)
    
```

### 4. Username Generator

```python
import random

full_name = input("Enter full name(your name should first name and last name with at least 3 letters): ").strip('"').lower()

if len(full_name.split()) < 2:
    print("Your name must have first name and last name")
else:
    first_name = full_name.split()[0]
    last_name = full_name.split()[1]
    if len(first_name) < 3 or len(last_name) < 3:
        print("Your first name and last name should have at least 3 letters.")
    else:
        username = first_name[:3] + last_name[-3:] + str(random.randint(10, 99))
        print(username)
        # Mamun Malitha
```

### 5. SMS Word Counter

```python
msg = input("Enter message(limit 160 characters): ").strip('"')
characters = len(msg)
words = len(msg.split())

if characters > 160:
    status = "Character Limit Exceeded."
else:
    status = "Within Limit"
print(f"Words = {words}, Characters = {characters}, Status = {status}")
#"This is a demo message."
```

### 6. Remove Duplicates from Contact List

```python
names = input("Enter the list of names: ").strip('"').split(",")
unique_names = []

for i in names:
    ch = i.strip()
    if ch not in unique_names:
        unique_names.append(ch)
# "Rahim, Karim, Rahim, Sultana"
print(f"\"{", ".join(unique_names)}\"")
```

### 7. Detect Suspicious Words in Text (Moderation)

```python
text = input("Enter Text: ").strip('"').lower().split()
banned = ['spam', 'scam', 'fake']
tag = True

for i in text:
    if i in banned:
        tag = False
        print("Warning: Suspicious content detected")
        break
# "This is a scam offer"
if tag:
    print("No suspicious content detected")
```

### 8. Invoice Code Formatter

```python
num = input("Enter invoice number(Your number should be of 5 digits): ")
if num.isdigit() == False:
    print("Your input is not a number.")
else:
    l = len(num)
    if l > 5:
        print("Your input number exceeds the 5 digit limit.")
    else:
        while len(num) < 5:
            num = "0" + num
        code = "INV" + num
        print(code)
```

### 10. Word Reversal for Data Encryption

```python
sentence = input("Enter Sentence: ").strip('"').split()
for i in range(0,len(sentence)):
    sentence[i] = sentence[i][::-1]
print(f"\"{" ".join(sentence)}\"")
# "AI is fun"
```
