### **Assignment 1: Temperature Check**

```python
temperature = float(input("Enter Temperature: "))
if temperature < 10:
    print("It's too cold!")
elif 10 <= temperature <= 25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

### **Assignment 2: ATM Withdrawal System**

```python
balance = float(input("Please enter your account balance: "))
withdraw = float(input("Please enter the amount you want to withdraw: "))

if withdraw > balance:
    print("Insufficient Balance")
elif withdraw < 0:
    print("Withdrawal amount cannot be negative.")
else:
    balance -= withdraw
    print(f"Transaction Successful")
    print(f"New balance: {balance:.2f}")
```

### **Assignment 3: Student Grade Calculator**

```python
marks = float(input("Enter marks(0 - 100): "))

if marks < 0 and marks > 100:
    print("Invalid input. Marks must be between 0 and 100")
if marks >= 90:
    print("A+")
elif marks >= 80:
    print("A")
elif marks >= 70:
    print("B")
elif marks >= 60:
    print("C")
else:
    print("Fail")
```

### **Assignment 4: Bus Ticket Discount**

```python
age = int(input("Enter your age: "))

if age < 0:
    print("Invalid age. Age cannot be negative.")
elif age < 5:
    print("Ticket Price: Free")
elif age >= 5 and age <= 18:
    print("Ticket Price: 50% Discount")
elif age >= 60:
    print("Ticket Price: 30% Discount")
else:
    print("Ticket Price: Full Price")
```

### **Assignment 5: Restaurant Bill with Discount**

```python
bill = float(input("Enter total amount: "))

if bill > 1000:
    bill -= bill * 0.1

print(f"Final bill: {bill:.2f}")
```

### **Assignment 7: Traffic Light System**

```python
color = input("Enter color: ")

if color == "red":
    print("Stop")
elif color == "yellow":
    print("Ready to go")
elif color == "green":
    print("Go")
else:
    print("Invalid color")
```

### **Assignment 8: Leap Year Checker**

```python
year = int(input("Enter year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100:
    print("Leap Year")
else:
    print("Not a leap year")
```

### **Assignment 9: Job Eligibility Checker**

```python
age = int(input("Enter age: "))
qualification = input("Enter qualification: ").lower()

if age >= 18 and qualification == "graduate":
    print("Eligible for Job")
else:
    print("Not Eligible")
```

### **Assignment 10: Electricity Bill Calculator**

```python
consumed = float(input("Enter the number of units consumed: "))

if consumed < 0:
    print("Invalid input. Units cannot be negative.")
elif consumed <= 100:
    bill = consumed * 5
elif consumed <= 200:
    bill = consumed * 7
else:
    bill = consumed * 10

print(f"Total payable amount: {bill}") 
```
