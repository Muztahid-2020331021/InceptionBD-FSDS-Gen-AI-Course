### **Assignment 1: Temperature Check**

```python
temperature = int(input("Enter Temperature: "))
if temperature < 10:
    print("It's too cold!")
elif 10 <= temperature <= 25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

### **Assignment 2: ATM Withdrawal System**

```python
balance = int(input("Please enter your account balance: "))
withdraw = int(input("Please enter the amount you want to withdraw: "))

if withdraw > balance:
    print("Insufficient Balance")
else:
    balance -= withdraw
    print(f"Transaction Successful")
    print(f"New balance: {balance}")
```

### **Assignment 3: Student Grade Calculator**

```python
marks = int(input("Enter marks(0 - 100): "))


if 90 <= marks <= 100:
    print("A+")
elif 80 <= marks <= 89:
    print("A")
elif 70 <= marks <= 79:
    print("B")
elif 60 <= marks <= 69:
    print("C")
else:
    print("Fail")
```

### **Assignment 4: Bus Ticket Discount**

```python
age = int(input("Please enter your age: "))
ticketPrice = 100

if age < 5:
    ticketPrice -= ticketPrice * 1
    discount = "Free"
elif 5 <= age <= 18:
    ticketPrice -= ticketPrice * 0.5
    discount = "50% Discount"
elif age >= 60:
    ticketPrice -= ticketPrice * 0.3
    discount = "30% Discount"
else:
    discount = "Full Price"

print(f"Ticket Price({discount}): {ticketPrice} Taka")
```

### **Assignment 5: Restaurant Bill with Discount**

```python
bill = int(input("Enter total amount: "))

if bill > 1000:
    bill -= bill * 0.1

print(f"Final bill: {bill}")
```

### **Assignment 7: Traffic Light System**

```python
color = input("Enter color: ")

if color == "red":
    print("Stop")
elif color == "yellow":
    print("Ready to go")
elif color == "green":
    print("Go")
else:
    print("Invalid color")
```

### **Assignment 8: Leap Year Checker**

```python
year = int(input("Enter year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100:
    print("Leap Year")
else:
    print("Not a leap year")
```

### **Assignment 10: Electricity Bill Calculator**

```python
consumed = int(input("Enter the number of units consumed: "))

if consumed <= 100:
    bill = consumed * 5
elif 101 <= consumed <= 200:
    bill = consumed * 7
else:
    bill = consumed * 10

print(f"Total payable amount: {bill}") 
```
