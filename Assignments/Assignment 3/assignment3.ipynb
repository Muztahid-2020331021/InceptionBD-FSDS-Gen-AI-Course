### 1. Inventory Price Calculator

```python
def calculate_inventory_summary(prices):
    total = sum(prices)
    average = total / len(prices)
    return total, average

def take_input():
    user_input = input("Please enter the list of prices: ").replace(',',' ').lstrip('[').rstrip(']').split()
    L = list(map(float, user_input))
    return L

def check_input(L):
    if len(L) == 0:
        print("Your list cannot be empty.")
        return False
    for i in L:
        if i < 0:
            print("Price cannot be negetive.")
            return False
    return True

# L = [120, 250, 399, 150]
L = take_input()
if check_input(L):
    print(calculate_inventory_summary(L))
```

### 2. Unique Visitor Tracker

```python
def count_unique_visitors(visitor_list):
    s = sorted(set(visitor_list))
    def fun():
        return ", ".join(s)
    ss = f"{len(s)} (The unique IDs are {fun()})"
    return ss

def take_input():
    user_input = input("Please enter the visitors' IDs: ").replace(',',' ').lstrip('[').rstrip(']').split()
    return user_input

def check_input(L):
    if len(L) == 0:
        print("There is no visitor.")
        return False
    return True

# L = [101, 205, 101, 310, 205, 550]
L = take_input()
if check_input(L):
    print(count_unique_visitors(L))
    
```

### 3. Product Category Counter

```python
def count_categories(products):
    d = {i:products.count(i) for i in products}
    return d

def take_input():
    user_input = input("Please enter the products' names: ").replace(',',' ').lstrip('[').rstrip(']').split()
    return user_input

def check_input(L):
    if len(L) == 0:
        print("There is no product.")
        return False
    return True

# L = ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]
L = take_input()
if check_input(L):
    print(count_categories(L))
```

### 4. Temperature Analytics

```python
def analyze_temperatures(temp_list):
    high_list = list(filter(lambda x: x > 38.0, temp_list))
    normal_list = list(filter(lambda x: x <= 38.0, temp_list))
    return high_list, normal_list

def take_input():
    user_input = input("Please enter the list of temperatures: ").replace(',',' ').lstrip('[').rstrip(']').split()
    L = list(map(float, user_input))
    return L

def print_output(temp_tuple):
    high_list, normal_list = temp_tuple
    print(f"High List(>38째C): {high_list}")
    print(f"Normal List(<=38째C): {normal_list}")
    print(f"Full Output Tuple: {temp_tuple}")

# L = [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]

L = take_input()
print_output(analyze_temperatures(L))
```

### 5. Menu Price Lookup

```python
"""
Input pattern:
How many items in your menu?:
4
Coffee 3.50
Tea 2.50
Muffin 4.00
Sandwich 8.50
Enter the name of the item you want to order: 
"Muffin"

"""

def get_price(menu, item):
    if item in menu.keys():
        return f"{menu[item]:.2f}"
    else:
        return "Item not found"

def take_input():
    menu = {}
    n = int(input("How many items in your menu?: "))
    for i in range(n):
        l = input().split()
        item_name = l[0]
        item_price = float(l[1])
        menu[item_name] = item_price
    item = input("Enter the name of the item you want to order: ")
    return menu, item

menu, item = take_input()
# menu = {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
# item = "Muffin"
print(get_price(menu,item))
```

### 6. Student Grading System

```python
"""
Input Pattern:
Number of Students:
4
Alice 85 
Bob 78 
Charlie 32 
David 69

"""

def calculate_grade(score):
    if score >= 80:
        return "A+"
    elif score >= 70:
        return "A"
    elif score >= 40:
        return "A-"
    else:
        return "F"

def grade_students(student_dict):
    d = {key: calculate_grade(val) for key, val in student_dict.items()}
    return d

def take_input():
    d = {}
    n = int(input("Number of Students: "))
    for i in range(n):
        l = input().split()
        student_name = l[0]
        student_mark = int(l[1])
        d[student_name] = student_mark
    return d

def check_input(d):
    for i,j in d.items():
        if j < 0 or j > 100:
            print("Invalid Marks")
            return False
    return True

# {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}

d = take_input()
if check_input(d):
    print(grade_students(d))
```

### 7. Discount Calculator

```python
"""
Input Pattern:
Number of Products:
3
Keyboard 80.00 20
Mouse 25.00 10
Monitor 300.00 5

"""


def apply_discounts(products):
    l = []
    for i in products:
        p = i[1] - i[1] * i[2] / 100
        l.append(p)
    return l

def take_input():
    li = []
    n = int(input("Number of products: "))
    for i in range(n):
        l = input().split()
        product_name = l[0]
        product_price = float(l[1])
        discount = float(l[2])
        li.append((product_name,product_price,discount))
    return li

def check_input(L):
    for i in L:
        if i[1] < 0.0 or i[2] < 0.0:
            print("Price and Discount cannot be negative.")
            return False
    return True
L = take_input()
if check_input(L):
    print(apply_discounts(L))
# L = [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]
# print(apply_discounts(L))
```

### 8. Hashtag Generator

```python
import string
def generate_hashtags(sentence):
    l = sentence.lower().split()
    return list(map(lambda x:'#' + x, l))

def take_input():
    sen = input("Enter the Sentence: ").strip('"')
    sen = remove_puntuations(sen)
    return sen

def remove_puntuations(sen):
    punctuations = string.punctuation
    for i in punctuations:
        if i in sen:
            sen = sen.replace(i, ' ')    
    return sen

def check_input(sen):
    l = sen.split()
    if len(l) == 0:
        print("You have entered an empty string.")
        return False
    return True


sen = take_input()

if(check_input(sen)):
    print(generate_hashtags(sen))
# sentence = "AI, Data! Science"
# print(generate_hashtags(sentence))
```

### 9. User Authentication Mock

```python
CREDENTIALS = {"alex_s": "securepwd1", "maria_t": "password101", "john_d": "johndoe123"}
def login(username, password):
    if username in CREDENTIALS.keys() and password == CREDENTIALS[username]:
        return "Login Successful"
    else:
        return "Invalid Credentials"
    
def take_input():
    username = input("Username: ").strip('"')
    password = input("Password: ").strip('"')
    return username, password

def check_input(username, password):
    if len(username) == 0:
        print("Username cannot be empty.")
        return False
    if len(password) < 8:
        print("Password should be at least 8 characters.")
        return False
    return True

username, password = take_input()
if(check_input(username, password)):
    print(login(username, password))
```

### 10. Shopping Cart Bill Calculator

```python
"""
Input Pattern:
Enter the number of items in the shop: 
4
Milk  3.00 
Bread 2.50 
Eggs 4.00 
Cheese 5.00
Enter the cart list:
Milk Bread Milk Cheese Bread

"""

def calculate_total(cart, prices):
    total = 0.0
    for i in cart:
        total += prices[i]
    return total

def take_input():
    n = int(input("Enter the number of items in the shop: "))
    d = {}
    for i in range(n):
        item_name, item_price = input().split()
        d[item_name] = float(item_price)
    cart = input("Enter the cart list: ").split()
    return cart, d

def check_input(cart, prices):
    if len(cart) == 0:
        print("Your cart is empty. Add something to your cart.")
        return False
    for i,j in prices.items():
        if j < 0.0:
            print("Price cannot be negative!")
            return False
    return True

cart, prices = take_input()
if check_input(cart, prices):
    print(f"Total bill amount is {calculate_total(cart, prices)}")

# {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}
# ["Milk", "Bread", "Milk", "Cheese", "Bread"]
```
